# Melting  / Pivot

```python
df = pd.DataFrame({'first' : ['John', 'Mary'],
                   'last' : ['Doe', 'Bo'],
                   'height' : [5.5, 6.0],
                   'weight' : [130, 150]})

df
```

```python
# Melt the dataframe
# to have one row per observation
# one column per variable

melted_df = df\
.melt(id_vars = ['first', 'last']) \
.sort_values(by = ['first', 'last', 'variable'])

melted_df
```

# Use case 1: Range retention

```python
# Define cohorts by the month of the first order

# Take the first order date per customer
cohorts_df = orders_df \
.groupby('customerid')['orderdate'] \
.agg('min') \
.pipe(pd.DataFrame)


# Customerid is in the index
# Create a column with the customerid
cohorts_df = cohorts_df.reset_index()


# Extract the month from the order date
def get_month(date):
    return(date.month)

cohorts_df['join_month'] = cohorts_df['orderdate'].apply(lambda x: get_month(x))


# Drop the date column
cohorts_df = cohorts_df.drop('orderdate', axis = 1)

# Each customer Id is associated with the month
# they first purchased (join_month) - that is also
# the cohort ID
cohorts_df.head()
```

```python
# Count cohort sizes

cohort_sizes_df = cohorts_df \
.groupby('join_month')['customerid'] \
.agg('count') \
.rename('cohort_size') \
.pipe(pd.DataFrame) \
.reset_index()

cohort_sizes_df
```

```python
# Cohort activity for each month
# One row per customer who came back at least once
# 'activity_index' months after activation

# Get the activity month
df = pd.merge(orders_df, cohorts_df,
         how = 'inner',
         left_on = 'customerid',
         right_on = 'customerid')


# Add the activity month and index
df['activity_month'] = df['orderdate'].apply(lambda x: get_month(x))
df['activity_index'] = df['activity_month'] - df['join_month']


# Remove activity for the first month
df = df[df.activity_index > 0]


# Make sure there is a max of one row per customer per month
# (customers who came back at least once in a month)
df = df[['customerid', 'join_month', 'activity_index']].drop_duplicates()

df.head()
```

```python
# Count the number of returning customers
# per cohort, per month

activity_size_df = df \
.groupby(['join_month', 'activity_index'])['customerid'] \
.agg('count') \
.rename('activity_size') \
.pipe(pd.DataFrame) \
.reset_index()

activity_size_df.head()
```

```python
# Calculate the retention rate
# per month after activation for each cohort

retention_df = pd.merge(cohort_sizes_df, activity_size_df,
        how = 'inner',
        left_on = 'join_month',
        right_on = 'join_month')


retention_df['retention_rate'] = round(retention_df['activity_size'] / retention_df['cohort_size'] * 100, 2)
retention_df.head()
```

See https://matplotlib.org/tutorials/colors/colormaps.html for valid cmaps

```python
retention_tbl = retention_df \
.pivot_table(values = 'retention_rate',
             index = ['join_month', 'cohort_size'], 
             columns = 'activity_index')

retention_tbl.style.background_gradient(cmap = 'Wistia')
```

```python
# Alternative option to render table
import seaborn as sns
sns.heatmap(retention_tbl, annot=True)
```

# Use case 2: top 5 customers per month

```python
orders_df.head(10)
```

```python
# 'Copy' the orders_df
df = orders_df[['orderdate', 'customerid', 'totalamount']]

# Add a month column and remove order date column
df['order_month'] = df['orderdate'].apply(lambda x: get_month(x))
df = df.set_index('order_month')
df = df.drop('orderdate', axis = 1)

df.head()
```

```python
# count and sum the purchases per month per customer
df1 = df \
.groupby(['order_month', 'customerid'])['totalamount'] \
.agg(['count','sum']) \
.sort_values(by = ['order_month', 'sum'], ascending = (True, False))

df1.head()
```

```python
# Rank all the customers per month
df1['custrank'] = df1['sum'] \
.groupby('order_month') \
.rank(method = 'dense', ascending = False)

df1.head(10)
```

```python
# Select only the top 5 each month
df1[df1['custrank'] <= 5] \
.rename(columns = {'count': 'n_orders', 'sum': 'total_amount'})
```

# Use case 3: Rolling count of orders per customer per date

```python
#df = orders_df[(orders_df['customerid'] == 7090) | (orders_df['customerid'] == 8069)] \

df = orders_df \
.groupby(['customerid', 'orderdate'])['orderid'] \
.agg('count') \
.pipe(pd.DataFrame) \
.rename(columns = {'orderid': 'n_orders'}) \
.reset_index()

df.head(10)
```

```python
# Create a DF with one row per day per cust and n of orders per day per cust
all_dates_df = pd.DataFrame({'date': pd.date_range('1/1/2004', periods = 366)})

all_dates_df.head()
```

```python
# Get unique customer IDs
unique_cust_df = df['customerid'] \
.drop_duplicates() \
.pipe(pd.DataFrame)

unique_cust_df.head()
```

```python
# All customerId / date combo
unique_cust_df['key'] = 0
all_dates_df['key'] = 0

all_ids_dates_df = pd.merge(unique_cust_df, all_dates_df, on = 'key') \
.drop('key', axis = 1) \
.sort_values(['customerid', 'date'])

all_ids_dates_df.head()
```

```python
# Add a column with the number of orders per cust, per day
df1 = pd.merge(left = all_ids_dates_df, right = df,
               how = 'left',
               left_on = ['customerid', 'date'], right_on = ['customerid', 'orderdate']) \
.drop('orderdate', axis = 1) \
.sort_values(['customerid', 'date'])

df1.head()
```

```python
# Change the nans into 0s
df1['n_orders'] = df1['n_orders'].apply(lambda x: 0 if np.isnan(x) else x)
df1.head(10)
```

```python
# Count rolling number of orders
df1['cumcount'] = df1 \
.groupby('customerid')['n_orders'] \
.cumsum()

df1.head()
```

```python
df1[(df1['customerid'] == 7090)].plot(x = 'date', y = 'cumcount', grid = True)
```
