```python
n_lines = 0

unique_businesses = set()

with open(file_path) as f:
    for line in f:
        line_dict = json.loads(line)
        unique_businesses.add(line_dict['id'])
        
        n_lines = n_lines + 1
        
len(unique_businesses)
```

```python
neigh_bus_dict = {}

with open(file_path) as f:
    for line in f:
        line_dict = json.loads(line)
        
        business_id = line_dict['id']
       
        # check if the neighborhood is in the line
        if line_dict['neighborhoods'] != None:
            
            # GEt the list of neighborhoods
            neighborhoods = line_dict['neighborhoods'].split(sep = ', ')
                
            #iterate over the neighs:
            for neighborhood in neighborhoods:
            
                if neighborhood not in neigh_bus_dict.keys():
                    neigh_businesses = set()
                    neigh_businesses.add(business_id)
                    neigh_bus_dict[neighborhood] = neigh_businesses
                
                else:
                    neigh_bus_dict[neighborhood].add(business_id)

neigh_bus_dict
```

```python
neigh_size_dict = {}

for k, v in neigh_bus_dict.items():
    neigh_size_dict[k] = len(v)

neigh_size_dict
```

```python
import pandas as pd
```

```python
neighborhoods = []
n_businesses = []

for k, v in neigh_size_dict.items():
    neighborhoods.append(k)
    n_businesses.append(v)
```

```python
df = pd.DataFrame({'Neighborhood': neighborhoods, 'n_businesses': n_businesses})
df.head()
```

```python
df['rank'] = df['n_businesses'].rank(ascending = False)
df = df.sort_values(by = 'rank')

df.head()
```
